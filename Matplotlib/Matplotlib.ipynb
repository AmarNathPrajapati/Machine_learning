```python
from matplotlib import style
style.use('ggplot')
x = [1,2,3]
y = [-1,2,-3]
x2 = [4,5,6]
y2 = [-4,5,-6]
# plt.plot(x,y)
# plt.plot(x2,y2)
# plt.plot(x,y,label='first')
plt.plot(x,y,label = 'first')
plt.plot(x2,y2,label = 'second')
plt.legend()
plt.title("Amar's data")
plt.ylabel("Y-axix")
plt.xlabel("X-axis")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x = [1,2,3]
y = [-1,2,-3]
x2 = [4,5,6]
y2 = [-4,5,-6]
# plt.plot(x,y)
# plt.plot(x2,y2)
# plt.plot(x,y,label='first')
plt.scatter(x,y,label = 'first')
plt.plot(x2,y2,label = 'second')
plt.legend()
plt.title("Amar's data")
plt.ylabel("Y-axix")
plt.xlabel("X-axis")
plt.show()
```

```python

from matplotlib import pyplot as plt
import numpy as np
 
 
# Creating dataset
a = np.array([22, 87, 5, 43, 56,
              73, 55, 54, 11,
              20, 51, 5, 79, 31,
              27])
 
# Creating histogram
fig, ax = plt.subplots(figsize =(5, 3))
ax.hist(a, bins = [0, 25, 50, 75, 100])
 
# Show plot
plt.show()
```
